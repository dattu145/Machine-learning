```python
df = pd.read_csv('exercise.csv').rename(columns={'per capita income (US$)':'income'})
```

```python
plt.scatter(df.year,df.income,c='r')
plt.xlabel('year')
plt.ylabel('income')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df.income)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(df.year,df.income)
plt.plot(df.year,reg.predict(df[['year']]),c ='r')
plt.show()
```
